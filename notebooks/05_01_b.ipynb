```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_csv("../data/groceries.csv")
```

```python
data.head()
```

```python
basket_sets = data.stack().str.get_dummies().groupby(level=0).sum()
```

```python
basket_sets.head()
```

```python
baskets_sets_bool = basket_sets.astype(bool)
apriori(baskets_sets_bool, min_support=0.02)
```

```python
baskets_sets_bool = basket_sets.astype(bool)
apriori(baskets_sets_bool, min_support=0.02, use_colnames = True)
```

```python
df_combo = basket_sets.astype(bool)

frequent_itemsets = apriori(df_combo, min_support=0.002, use_colnames = True)

frequent_itemsets["length"] = frequent_itemsets["itemsets"].apply(lambda x: len(x))

frequent_itemsets
```

```python
frequent_itemsets[(frequent_itemsets["length"] >=3)]
```

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.9)
rules.head()
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.5)
rules.head()
```

```python
rules[(rules["lift"]>=5) & (rules["confidence"]>=0.8)]
```
